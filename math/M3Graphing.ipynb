```python
#Creating x and y variables to graph
x = (df_construction['Month-Year'])
y = (df_construction['Total Construction'])
```

```python
fig,ax = plt.subplots()
ax.plot(x, y, linewidth=1.5, color = 'k')
ax.grid(False)

""" Enhance axes object """
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Date',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Total Construction Spending',fontsize='14',fontname='Times New Roman')
ax.set_xticks(x[::5])
plt.xticks(rotation=45)
ax.yaxis.set_major_formatter('${x:1.0f}')

""" Enhance figure object """
fig.set_size_inches(18,6)
fig.suptitle("Total 2021 Construction Spending",fontsize='18',fontname='Times New Roman')
fig.savefig('Total 2021 Construction Spending.jpg', dpi=2000)
plt.show()
```
